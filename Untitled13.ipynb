Function 1. Eliminate implication


```python
def EliminateImplication(kb):

    j = kb.find("⇒")


    if j != -1:
        # Split the string into two parts based on the implication symbol
        FristPart = kb[:(j)]
        SecPart = kb[(j + 2):]


        new_expression = "¬" + FristPart +"∨" + SecPart

        return new_expression


    return kb

expression = "(p ⇒ q)"
result = EliminateImplication(expression)
print(result)
```

Function 2. Move negation inward (Demorgan Law)

```python
def move_negation_inward(sentence):
    new_sentence = ""
    i = 0
    flag1=0
    flag2=0
    while i < len(sentence):
        if sentence[i] == '¬':
            i += 1
            if sentence[i] == '(':
                new_sentence += '¬'
                i += 1
                while i < len(sentence) and sentence[i] != ')':
                    if sentence[i] == '∨':
                        new_sentence += '∧'
                        new_sentence += '¬'
                    elif sentence[i] == '∧':
                        new_sentence += '∨'
                        new_sentence += '¬'
                    else:
                        new_sentence += sentence[i]
                    i += 1
            elif sentence[i] == '∃':
              new_sentence +='∀'
              i+=1
              while not sentence[i].isalpha():
                i+=1
              new_sentence += sentence[i]
              i+=1
              new_sentence += '¬'
              i+=1
            elif sentence[i] == '∀':
              new_sentence +='∃'
              i+=1
              while not sentence[i].isalpha():
                i+=1
              new_sentence += sentence[i]
              i+=1
              new_sentence += '¬'
              i+=1



        else:

            new_sentence += sentence[i]
            i += 1
    new_sentence = new_sentence.replace(')', '')

    return new_sentence

print(move_negation_inward("¬( p ∧ q)"))
```

Function 3. Remove double-not.

```python
def RemoveDoubleNot(kb):
    j = kb.find("¬¬")
    if j != -1:
        kb = kb.replace("¬¬", "")
    return kb

print(RemoveDoubleNot('¬¬ p → p'))
```

Function 4. Standardize variable scope.

```python
import random
import string
def replace_char_from_index(string, char, replacement, start_index):
    new_string = string[:start_index]
    for i in range(start_index, len(string)):
        if string[i] == char:
            new_string += replacement
        else:
            new_string += string[i]
    return new_string
def standardize_variables(statement):
    new_sentence = ""
    i = 0
    while i < len(statement):
        if statement[i] == '∃' or statement[i] == '∀':
            i += 1
            while not statement[i].isalpha():
                i += 1
            if statement[i]  not in new_sentence:
                new_sentence += statement[i]
                i+=1
            else:

                random_char = random.choice(string.ascii_letters)
                while random_char in new_sentence:
                    random_char = random.choice(string.ascii_letters)
                new_sentence += random_char
                statement=replace_char_from_index(statement,statement[i],random_char,i)
                i += 1


        else:
         i += 1
    return statement

# Example usage:
statement = '(∀x P( x)) ∨ (∃x Q( x))'
print(standardize_variables(statement))
```

Function 6. Skolemization for existential quantifiers.

```python
def skolemization(formula):
    symbol = '∃'
    result = ''
    i = 0
    counter = 1


    while i < len(formula) - 1:
        if formula[i] == symbol and formula[i+1] not in result:
            result += formula[i+1]
            formula = formula.replace(formula[i+1], '', 1)
        else:
            i += 1

    formula = formula.replace(symbol, '')


    for char in result:
        if char in formula and char.isalpha():
            FX = "F" + str(counter) + "(x)"
            formula = formula.replace(char, FX,10000)
            counter += 1

    return formula

print(skolemization("∀x ∃y  ∃z P(x) ∨ Q(y) p(y) p(z) "))
```

Function 7. Eliminate universal quantifiers.

```python
def EliminateQuantifier(kb):

    NewKb = ""

    i = 0

    while i < len(kb):

        if kb[i] == "∀" or kb[i] == "∃":
            # Skip the quantified variable
            i += 1
            # Iterate over the quantified variable until a non-alphabetic character is encountered
            while i < len(kb) and kb[i].isalpha():
                i += 1

            continue
        else:

            NewKb += kb[i]

        i += 1

    return NewKb

# Test the function with an example input
print(EliminateQuantifier("∀x P(x) ∨ Q(F(x)) → P(x) ∨ Q(F(x))"))
```

Function 8. Convert to conjunctive normal form

```python
def Distributive(kb):
    bract = ""
    j = 0
    lis_brack = []
    distr_list = []
    st = ""

    for char in kb:
        if char == "(":
            j += 1

        if j == 1:
            bract += char

        if char == ")":
            j = 0
            lis_brack.append(bract)
            bract = ""
            continue

        if char not in ("(", ")") and char != " ":
            lis_brack.append(char)



    for i, char in enumerate(lis_brack):
        if char == "∨":

            left_term = lis_brack[i - 1]
            right_term = lis_brack[i + 1]

            # Append the distributed terms to the list
            distr_list.append("(" + left_term + " ∨ " + right_term + ")")
            distr_list.append("(" + left_term + " ∨ " + lis_brack[i+3] + ")")



    st = ' ∧ '.join(distr_list)
    return st



kb = "p ∨ (q ∧ r)"
result = Distributive(kb)
print(result)
```
